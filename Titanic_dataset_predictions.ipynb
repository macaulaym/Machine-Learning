Data cleaning

Dropping missing values

```python
titanic = titanic.drop('Cabin', axis=1)
titanic
```

```python
# removing all rows with missing values
titanic_removed_missing = titanic.dropna()
titanic_removed_missing
```

```python
titanic_removed_missing.info()
```

```python
# replcing missing data by the average
titanic_replacing_missing_with_median = titanic.fillna(titanic.median())
titanic_replacing_missing_with_median
```

```python
titanic_replacing_missing_with_median.info()
```

Preparing data for machine learning

```python
# Selecting columns that we will use as features for the machine learning model
X = titanic_removed_missing[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
# Choosing the lables column (what we want the machine learning model to predict)
y = titanic_removed_missing['Survived']
```

```python
X
```

```python
y
```

```python
# Converting the categorical columns into dummy variables (a column for each category where one of them is 1 and the others are 0).
# e.g for the Embarked column, there are three possible values, S, C, Q, we create 3 new columns for them, and if the value is S then it will be 1 in the new S column
# and 0 in the other two columns
X = pd.get_dummies(X)
X
```

```python
# Splitting the data into a training set and a test set, were we training the modlel we train it on the training set and evaluate it on the unseen test set.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

```python
X_test
```

Machine Learning

1. Logistic regression

```python
# import the algorithm
from sklearn.linear_model import LogisticRegression
```

```python
# initialize a variable for the machine learning model (here the breckets are empty meaning that we are using the default parameters for the algorithm)
lr = LogisticRegression()
```

```python
# training the model (to update the internal weights to minimize the error on the training set)
lr.fit(X_train, y_train)
```

```python
# caluclating the accuracy score of the training set
lr.score(X_train, y_train)
```

```python
# caluclating the accuracy score of the test set (this is the number that really matters, the higher the test accuracy the better the model is)
lr.score(X_test, y_test)
```

```python
# using the machine learning model to make predictions of whether the people in the X_test set have survived on not
y_pred = lr.predict(X_test)
```

```python
# printing the true y values and the predicted y values to compare them. The more similar the values are the better.
[y_test.values, y_pred]
```

Random forest

```python
# import the algorithm
from sklearn.ensemble import RandomForestClassifier
```

```python
# initialize a variable for the machine learning model (here the breckets are empty meaning that we are using the default parameters for the algorithm)
rf = RandomForestClassifier()
```

```python
# training the model (to update the internal weights to minimize the error on the training set)
rf.fit(X_train, y_train)
```

```python
# caluclating the accuracy score of the training set
rf.score(X_train, y_train)
```

```python
# caluclating the accuracy score of the test set (this is the number that really matters, the higher the test accuracy the better the model is)
rf.score(X_test, y_test)
```

```python
# using the machine learning model to make predictions of whether the people in the X_test set have survived on not
y_pred = rf.predict(X_test)
```

```python
# Save the first decision tree to a file called tree.png
estimator = rf.estimators_[0]

from sklearn.tree import export_graphviz
# Export as dot file
export_graphviz(estimator, out_file='tree.dot', 
                feature_names = X.columns,
                class_names = ['Died', 'Survived'],
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])
```

```python
# Trying again with a max deptth of 5
rf = RandomForestClassifier(max_depth=5)
```
